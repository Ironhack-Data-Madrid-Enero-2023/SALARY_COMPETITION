```python
def get_value_counts(df):
    """
    Devuelve los value_counts de todas las columnas de un DataFrame.
    """
    value_counts_dict = {}
    for column in df.columns:
        value_counts_dict[column] = df[column].value_counts()
    return value_counts_dict
```

```python
print(get_value_counts(salaries)) # Según esto, tenemos varias columnas que, en principio están bien, pero otras debemos
# transformarlas, como por ejemplo job title, que habría que homogeneizarla y hacer un get dummies para codificar cada
# puesto de trabajo.
```

Vamos a ver ahora de qué columnas podemos prescindir; para ello, en primera instancia vamos a tratar de aplicar el sentido común y, más adelante, veremos si se puede/debe prescindir de alguna variable más.

- Las columnas que pienso que podríamos prescindir de ellas porque no van a tener relación con los salarios, son las siguientes: work_year, employment_type, salary (no hay uniformidad, ya que cada salario está expresado en diferentes divisas), salary_currency y remote_ratio.

```python
# Vamos a dropearlas, pues:
salaries = salaries.drop(['work_year', 'employment_type', 'salary', 'salary_currency', 'remote_ratio'], axis=1)
```

```python
salaries.head()
```

Antes de hacer get-dummies sobre lo no numérico, vamos a homogeneizar las columnas employee_residence y company_location. El criterio que vamos a seguir va a ser quedarnos con los 10 valores (países en este caso) que más se repiten y el resto lo englobamos en Others.

```python
paises = salaries.employee_residence.value_counts()[:10]
```

```python
salaries['employee_residence'] = salaries['employee_residence'].apply(lambda x: x if x in paises else 'OTHER')
```

```python
salaries.employee_residence.value_counts()
```

```python
paises_2 = salaries.company_location.value_counts()[:10]
```

```python
salaries['company_location'] = salaries['company_location'].apply(lambda x: x if x in paises else 'OTHER')
```

```python
salaries.company_location.value_counts()
```

```python
salaries = pd.get_dummies(salaries, columns=['experience_level','job_title', 'employee_residence', 'company_location', 'company_size'], drop_first=True)
```

```python
salaries.head()
```

```python
# Como podemos ver en la matriz de correlaciones, no hay fuertes relaciones entre las variables y el salario en USD.
# Lo más relacionado es "employee_residence_US" y "company_location_US".

plt.figure(figsize=(15, 10))

sns.set(style='white')

mask=np.triu(np.ones_like(salaries.corr(), dtype=bool))

cmap=sns.diverging_palette(0, 10, as_cmap=True)


sns.heatmap(salaries.corr(),
           mask=mask,
          cmap=cmap,
          center=0,
          square=True,
          annot=True,
          linewidths=0.5,
          cbar_kws={'shrink': 0.5});
```

Vamos ahora con la parte de Machine Learning:

```python
X = salaries.drop('salary_in_usd', axis=1)

y = salaries.salary_in_usd
```

```python
scaler = StandardScaler()
```

```python
X_norm = scaler.fit_transform(X)
```

```python
#X_norm = pd.DataFrame(X_norm, columns=salaries.columns)

#X_norm.head()
```

```python
# Vamos a entrenar:

X_train, X_test, y_train, y_test  = tts(X_norm, y, train_size=0.8, test_size=0.2, random_state=42)

X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
#pip install lazypredict
```

```python
lazy = LazyRegressor()
```

```python
lazy.fit(X_train, X_test, y_train, y_test)
```

```python
gamma_reg = GammaRegressor()
gamma_reg.fit(X_train, y_train)
```

```python
y_pred = gamma_reg.predict(X_test)
```

```python
mse = sq(y_test, y_pred, squared=False)
mse
```
