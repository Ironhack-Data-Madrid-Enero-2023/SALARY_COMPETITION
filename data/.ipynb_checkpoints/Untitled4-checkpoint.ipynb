```python
import pandas as pd
```

```python
traintr=pd.read_csv('salaries_data.csv')
```

```python
traintr.head()
```

```python
traintr = traintr.drop(['salary', 'salary_currency'], axis=1)
```

```python
traintr.head()
```

```python
traintr.info()
```

```python
traintr.experience_level.value_counts()
```

```python
dictio = {'EN':0 , 'MI':1, 'SE':2,'EX':3}

traintr.experience_level=traintr.experience_level.apply(lambda x: dictio[x])
```

```python
traintr.employment_type.value_counts()
```

```python
dictio = {'FL':0 , 'CT':1, 'PT':2,'FT':3}

traintr.employment_type=traintr.employment_type.apply(lambda x: dictio[x])
```

```python
traintr.job_title.value_counts()
```

```python
dictio = {'Data Analyst': 1,
    'Data Engineer': 2,
    'Data Scientist': 3,
     'Other': 0 }

traintr['job_title_numeric'] = traintr['job_title'].apply(lambda x: dictio[x] if x in dictio else dictio['Other'])
```

```python
traintr=traintr.drop('job_title',axis=1)
```

```python
traintr
```

```python
from sklearn.preprocessing import LabelEncoder  
```

```python
le = LabelEncoder()
```

```python
traintr["employee_residence"] = le.fit_transform(traintr["employee_residence"]) # chequear si se puede agrupar por continente y luego asignarle un numero
```

```python
traintr["company_location"] = le.fit_transform(traintr["company_location"])
```

```python
dictio = {'S':0 ,'L':1, 'M':2}

traintr.company_size=traintr.company_size.apply(lambda x: dictio[x])
```

```python
traintr.head()
```

```python
X=traintr[['work_year','experience_level','employment_type','job_title_numeric','employee_residence','remote_ratio','company_location','company_size']]
y=traintr['salary_in_usd']
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scalder=StandardScaler()
```

```python
traintr=scalder.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split as tts
```

```python
X_train, X_test, y_train, y_test  = tts(X, y, train_size=0.8, test_size=0.2, random_state=42)

X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
from lazypredict.Supervised import LazyRegressor
```

```python
lazy=LazyRegressor()
```

```python
lazy.fit(X_train, X_test, y_train, y_test)
```

```python
import lightgbm as lgb
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

```python
model = lgb.LGBMRegressor()
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
mse = mean_squared_error(y_test, y_pred, squared=False)
mse
```
