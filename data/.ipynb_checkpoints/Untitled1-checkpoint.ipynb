```python
import pandas as pd
```

```python
train_t=pd.read_csv('salaries_data.csv')
```

```python
train_t.head()
```

```python
train_t = train_t.drop(['salary', 'salary_currency'], axis=1) # me cargo estas columnas porque ya tengo 'salary_in_usd'
```

```python
train_t
```

```python
train_t.info()
```

```python
train_t = train_t.drop(['employment_type'], axis=1)
```

```python
train_t
```

```python
train_t.experience_level.value_counts()
```

```python
train_t.job_title.value_counts()
```

```python
train_t.employee_residence.value_counts()
```

```python
train_t.company_location.value_counts()
```

```python
train_t.company_size.value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder # uso label encoder para las categóricas que tienen muchos valores ya que si usara get_dummies me generaría muchas columnas.
```

```python
le = LabelEncoder()
```

```python
train_t["job_title"] = le.fit_transform(train_t["job_title"])
train_t["employee_residence"] = le.fit_transform(train_t["employee_residence"])
train_t["company_location"] = le.fit_transform(train_t["company_location"])
```

```python
train_t.info()
```

```python
dummies_exp = pd.get_dummies(train_t['experience_level'], prefix='exp') # uso get dummies para las categoricas que tienen pocos valores
dummies_size = pd.get_dummies(train_t['company_size'], prefix='size')
```

```python
train_t= pd.concat([train_t, dummies_exp, dummies_size], axis=1)
```

```python
train_t.drop(['experience_level', 'company_size'], axis=1, inplace=True)
```

```python
train_t
```

```python
train_t.describe()
```

```python
import pylab as plt
import seaborn as sns
import numpy as np

%matplotlib inline    
```

```python
plt.figure(figsize=(15, 10))      # pinto un grafico de correlacion para ver las relaciones de las Xs y luego de las Xs con la Y

sns.set(style='white')

mask=np.triu(np.ones_like(train_t.corr(), dtype=bool))

cmap=sns.diverging_palette(0, 10, as_cmap=True)


sns.heatmap(train_t.corr(),
           mask=mask,
          cmap=cmap,
          center=0,
          square=True,
          annot=True,
          linewidths=0.5,
          cbar_kws={'shrink': 0.5});
```

```python
train_t = train_t.drop(['company_location'], axis=1) # me cargo company location porque esta X está muy correlada con employee_residence por lo que puede que duplique a la hora de netrenar la maquina

# además la quito también porque a la hora de la correlación es la que menos de las dos lo está con la y (salary_in_usd) 
```

```python
plt.figure(figsize=(15, 10))      

sns.set(style='white')

mask=np.triu(np.ones_like(train_t.corr(), dtype=bool))

cmap=sns.diverging_palette(0, 10, as_cmap=True)


sns.heatmap(train_t.corr(),
           mask=mask,
          cmap=cmap,
          center=0,
          square=True,
          annot=True,
          linewidths=0.5,
          cbar_kws={'shrink': 0.5});
```

```python
X=train_t[['employee_residence','exp_EN','exp_EX','exp_MI','exp_SE']]  
y=train_t['salary_in_usd']

# Una vez que tengo las correlaciones escojemos las columnas más correlacionadas para ello defino la X y la Y
```

```python
from sklearn.preprocessing import StandardScaler # normalizamos los datos 
```

```python
scaler = StandardScaler()
```

```python
X_scaled = scaler.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split as tts # dividimos para luego entrenar el modelo
```

```python
X_train, X_test, y_train, y_test  = tts(X, y, train_size=0.8, test_size=0.2, random_state=42)

X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
#%pip install lazypredict
```

```python
from lazypredict.Supervised import LazyRegressor
```

```python
lazy=LazyRegressor()
```

```python
lazy.fit(X_train, X_test, y_train, y_test)
```

```python
help(lazy)
```

```python
from sklearn.linear_model import PassiveAggressiveRegressor
from sklearn.metrics import r2_score, mean_squared_error as sq
```

```python
model = PassiveAggressiveRegressor()
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
r2 = r2_score(y_test, y_pred) 
r2
```

```python
mse = sq(y_test, y_pred,squared=False)
mse
```
