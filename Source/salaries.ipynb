```python
df = pd.concat([df_salary, df_test])
```

```python
df
```

```python
#Jerarquizacion de la experiencia en los puestos de trabajo
experience_level = {'EX': 1 , 'SE': 2, 'MI': 3, 'EN': 4}
df.experience_level=df.experience_level.apply(lambda x: experience_level[x])
```

```python
df
```

```python
types = {'FL': 1, 'CT': 2, 'PT': 3, 'FT': 4}
employment_type = []
for e in df.employment_type:
    if e in types:
        employment_type.append(types[e])
```

```python
df['employment_type'] = employment_type
```

```python
le = LabelEncoder()

# Aplicamos Label Encoding a la columna 'frutas'
df['company_size'] = le.fit_transform(df['company_size'])
```

```python
df
```

```python
df['company_location'] = le.fit_transform(df.company_location)
df['job_title'] = le.fit_transform(df.job_title)
```

```python
df
```

```python
df.drop(columns = ['employee_residence'], inplace = True)
```
